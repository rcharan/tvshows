```python
to_do_1 = assessement['Parser 1']
def parser_1(show):
    out_list = []
    for i in range(len(show)):
        if show[i] in [':', '-', '(']:
            out_list.append(show[:i])
    return out_list
```

```python
good_shows = find_urls_with_parser(to_do_1, parser_1)
```
